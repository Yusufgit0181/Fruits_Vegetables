```python
import pandas as pd
import numpy as np
import requests
import json
from pandas import json_normalize
from seperate_file_password import database_password

url = "https://www.ec.europa.eu/agrifood/api/fruitAndVegetable/prices?"

def get_data(source):
    data = requests.get(source)
    if data.status_code == 200:
        print("Data obtained")
    else:
        raise Exception(f"Non-success status code: {data.status_code}")
    
    data = data.content.decode('utf-8')
    data = json.loads(data)
    data = pd.DataFrame(data)

    # Display the DataFrame
    return data
data = get_data(url)
data
```

Data Validation

```python
# These are the column we expect
expected_columns = ["memberStateCode", "memberStateName", "beginDate", "endDate", "weekNumber"	,"price","unit"	,"product", "variety", "description"]

# Check if the actual columns match the expected columns
missing_columns = [col for col in expected_columns if col not in data.columns]
extra_columns = [col for col in data.columns if col not in expected_columns]


assert len(missing_columns) == 0, f"Missing columns: {missing_columns}"
assert len(extra_columns) == 0, f"Unexpected columns found: {extra_columns}"

print("All columns are present and correct")
```

```python
#This is dropped as it does not tell us anything and just sums all the info from the other columns
data = data.drop(["description","memberStateCode"], axis = 1)
data
```

```python
data.info()
```

```python
data["beginDate"] = pd.to_datetime(data["beginDate"], format='%d/%m/%Y').dt.date
data["endDate"] = pd.to_datetime(data["endDate"], format='%d/%m/%Y').dt.date

data
```

```python
data.info()
```

```python
# One thing I prefer is to make the data as clean as possible, as early as possible.
# We're gonna want to make the unit collumn state the metric being used such that future calculations for that column will be easier.

data.rename(columns={"unit": "unit/KG"}, inplace= True)
data["unit/KG"] = data["unit/KG"].str.replace("KG", "") 
data["unit/KG"] = data["unit/KG"].astype(int)

data.rename(columns={"price": "price in €"}, inplace= True)
data["price in €"] = data["price in €"].str.replace("€", "") 
data["price in €"] = data["price in €"].astype(float)
```

```python
data.info()
```

```python
# Let's help ourselves and make an Product_ID column by combining product and variety
# If the below string is found it won't tell us more about the product, hence why we don't need it and the product itself becomes unique.
data.loc[data['variety'].str.contains("All types and varieties", na=False), 'variety'] = np.nan
data["ID"] = data["product"] + "_" + data["variety"]
data["ID"].fillna(data["product"], inplace=True)
data
```

New data Check

```python
import psycopg2
from sqlalchemy import create_engine

conn = psycopg2.connect(database="postgres",
                        host="localhost",
                        user="postgres",
                        password=database_password,
                        port="5432")


def sql_connect(db_url):
    return create_engine(db_url)

engine = sql_connect(f'postgresql://postgres:{database_password}@localhost/postgres')

table_name = "prices"


# We store the latest date (which indicates new data) in a variable and query our DB to check if records exist
# If these records exist it must mean that this is not new data
last_observation = data["endDate"].max().strftime('%Y-%m-%d')



# Here we check if the last observation in the 'new' data is bigger than the latest date in the db 
df = pd.read_sql_query(f"SELECT * FROM {table_name} WHERE \"endDate\" > '{last_observation}';", engine)

df


```

Dump data in DB

```python

if df.empty:
    print("There is no new data")
else:
    data.to_sql(table_name, engine, if_exists='append', index=False) 
    print("New data found. Appended to DB")
    
```
